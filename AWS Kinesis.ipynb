## Kinesis Video Stream

- Securely stream video from devices to AWS for further processing.
- Automatically provisions and scales infrastructure to ingest video data from any number of devices.
- The video data is stored, encrypted, indexed and made available through APIs.

### Use Cases

- Smart Home (Baby Monitoring, Camera Enabled Doorbell)
- Smart City (CCTV Live Stream Analysis)
- Smart Automation (Temperature Monitoring for Industrial Equipment using thermal imagery)


## Kinesis Data Stream

- Scalable and Durable real-time data streaming service that can capture gigabytes of data from multiple sources.
- The data collected is available within a few milliseconds to enable real-time analysis.
- Data sources include website clickstreams, financial transactions, database event streams, social media feeds, IT logs and location tracking events.

### Use Cases

- Real-Time Dashboards
- Real-Time Anomaly Detection
- Real-Time Fraud Detection
- Dynamic Pricing

## Kinesis Data Firehose

- Data Firehose loads the streaming data into data lakes, data stores and analytics services.
- It can capture, transform and deliver data to aws storage services, http endpoints and service providers like MongoDB and Splunk.
- It loads the streaming data into the storage medium within 60 seconds of receiving it.

### Use Cases

- IoT Analytics
- Clickstream Analytics
- Log Analytics
- Security Monitoring

## Differences between Kinesis Data Stream and Kinesis Data Firehose

| Kinesis Data Stream | Kinesis Data Firehose |
| ------------------- | --------------------- |
| Real-Time Analytics | Near Real-Time Analytics|
| Data available within milliseconds | Data can be available within 60 seconds |
| Manual configuration of shards is required for scaling | Features automated scaling |
